```python
import pandas as pd

df = pd.read_csv("student_scores.csv")
print("Shape before cleaning:", df.shape)
df.head()
```

```python
df.isnull().sum()
```

```python
# Show the actual rows that have missing values
df[df.isnull().any(axis=1)]
```

```python
# Fill missing values with the column mean
# (a simple, common approach for small numeric datasets like this one)
df["StudyHours"] = df["StudyHours"].fillna(df["StudyHours"].mean())
df["Score"] = df["Score"].fillna(df["Score"].mean())

print("Missing values after filling:")
print(df.isnull().sum())
```

```python
print("Number of duplicate rows:", df.duplicated().sum())
df[df.duplicated()]
```

```python
df = df.drop_duplicates().reset_index(drop=True)
print("Shape after removing duplicates:", df.shape)
print("Duplicate rows remaining:", df.duplicated().sum())
```

```python
df.describe()
```

```python
print("Mean study hours:", round(df["StudyHours"].mean(), 2))
print("Mean score:", round(df["Score"].mean(), 2))
print("Max score:", df["Score"].max())
print("Min score:", df["Score"].min())
print("Correlation between StudyHours and Score:", round(df["StudyHours"].corr(df["Score"]), 2))
```

```python
df.to_csv("student_scores_cleaned.csv", index=False)
print("Cleaned dataset saved as student_scores_cleaned.csv")
```
