Tasks: Create scatter plots, bar charts and line charts using Matplotlib.

## 1. Load the Cleaned Dataset

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("student_scores_cleaned.csv")
df.head()
```

## 2. Scatter Plot
A scatter plot helps us see the relationship between study hours and scores —
does studying more actually lead to a higher score?

```python
plt.figure(figsize=(8, 5))
plt.scatter(df["StudyHours"], df["Score"], color="royalblue", edgecolor="black")
plt.title("Study Hours vs Score")
plt.xlabel("Study Hours")
plt.ylabel("Score")
plt.grid(True, alpha=0.3)
plt.savefig("scatter_study_vs_score.png", dpi=150, bbox_inches="tight")
plt.show()
```

## 3. Bar Chart
Let's group students into study-hour ranges (bins) and compare their average scores.

```python
bins = [0, 2, 4, 6, 8, 10]
labels = ["0-2", "2-4", "4-6", "6-8", "8-10"]
df["HoursGroup"] = pd.cut(df["StudyHours"], bins=bins, labels=labels)

avg_scores = df.groupby("HoursGroup", observed=True)["Score"].mean()

plt.figure(figsize=(8, 5))
avg_scores.plot(kind="bar", color="seagreen", edgecolor="black")
plt.title("Average Score by Study Hours Group")
plt.xlabel("Study Hours Range")
plt.ylabel("Average Score")
plt.xticks(rotation=0)
plt.grid(True, axis="y", alpha=0.3)
plt.savefig("bar_avg_score_by_group.png", dpi=150, bbox_inches="tight")
plt.show()
```

## 4. Line Chart
Sorting by study hours and plotting as a line helps show the overall upward trend.

```python
sorted_df = df.sort_values("StudyHours")

plt.figure(figsize=(8, 5))
plt.plot(sorted_df["StudyHours"], sorted_df["Score"], marker="o", color="darkorange")
plt.title("Score Trend as Study Hours Increase")
plt.xlabel("Study Hours")
plt.ylabel("Score")
plt.grid(True, alpha=0.3)
plt.savefig("line_score_trend.png", dpi=150, bbox_inches="tight")
plt.show()
```

## Summary
Today I created three types of charts using Matplotlib:
- A **scatter plot** to see the raw relationship between study hours and scores
- A **bar chart** comparing average scores across study-hour groups
- A **line chart** showing the overall upward trend

All three charts confirm what we'd expect: more study hours generally lead to higher scores —
a good sign that Linear Regression (coming up on Day 7-8) should work well on this data.
