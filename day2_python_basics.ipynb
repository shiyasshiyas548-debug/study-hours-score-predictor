```python
# Variables and Data Types
name = "Shiyas"           # text (string)
age = 21                  # whole number (integer) - change to your age
study_hours = 6.5         # decimal number (float)
is_studying = True        # True/False (boolean)

print(name, age, study_hours, is_studying)
print(type(name), type(age), type(study_hours), type(is_studying))
```

```python
# Operators
a = 10
b = 3

print("Addition:", a + b)
print("Subtraction:", a - b)
print("Multiplication:", a * b)
print("Division:", a / b)
print("Is a greater than b?", a > b)
```

```python
# Functions
def greet_student(student_name):
    return f"Hello, {student_name}! Ready to study?"

message = greet_student("Shiyas")
print(message)
```

```python
# Simple program: study hours -> predicted effort level
def predict_effort(hours):
    if hours < 2:
        return "Low effort"
    elif hours < 5:
        return "Moderate effort"
    else:
        return "High effort"

study_days = [1, 3, 6, 2, 8]

for hours in study_days:
    print(f"{hours} hours -> {predict_effort(hours)}")
```
